# Class and Objects
Python is an object oriented programming language.

Almost everything in Python is an object, with its properties and methods.

A Class is like an object constructor, or a "blueprint" for creating objects.

```python
class sample:
  def show(self):
    print("Welcome")

# Objects
s1=sample()
s1.show()
```

```python
class Person:
    age = 10

    def greet(self):
        print('Hello')
```

```python
# Create Objects
```

```python
s1=Person()
print(s1.age)
s1.greet()
```

```python
class pc:
  def config(self):
    print("i7","8gb","1Tb")

pc1=pc()
pc2=pc()
pc1.config()
pc2.config()

a="abc"
print(type(a))
print(type(pc1))
```

## _init_ methods: Special Method



```python
class computer:

  def __init__(self,cpu,ram):
    self.cpu=cpu
    self.ram=ram


  def config(self):
    print("Config:\t", self.cpu,self.ram)


pc1=computer("i5","8gb")
pc2=computer("i7","16gb")

pc1.config()
pc2.config()


```

```python
class cal:
  def __init__(self,x,y):
    self.x=x
    self.y=y


  def add(self):
    print(self.x+self.y)

s1=cal(10,20)
s1.add()
print(s1)



```

```python
# Display Welcome n times

class sample:
  def __init__(self,n):
    self.n=n

  def show(self):
    for i in range(0,self.n):
      print("Welcome")

# Objects
n=int(input("Enter n=\t"))
s1=sample(n)
s1.show()
```

```python
class details:

  def __init__(self):   # Default Arguements
    self.name="ravi"
    self.age=30


  def show(self):
    print("Name:",self.name)
    print("Age:",self.age)



c1=details()
c2=details()

print("Before")
c1.show()
c2.show()

# Change values
c2.name="Akash";
c2.age="20"

print("After")
c1.show()
c2.show()
```

```python
# Update through Function

class details:

  def __init__(self):   # Default Arguements
    self.name="ravi"
    self.age=30

  def update(self,name,age):
    self.name=name
    self.age=age

  def show(self):
    print("Name:",self.name)
    print("Age:",self.age)


c1=details()
c2=details()

print("Before")
c1.show()
c2.show()

# Change values
c1.update("Akash",20)

print("After")
c1.show()
c2.show()
```

```python
# Compare the age
class details:

  def __init__(self):   # Default Arguements
    self.name="ravi"
    self.age=30

  def update(self,name,age):
    self.name=name
    self.age=age

  def show(self):
    print("Name:",self.name)
    print("Age:",self.age)

  def compare(self,other):
    if(self.age==other.age):
      print("Same age")
    else:
      print("diffrent age")

c1=details()
c2=details()

c1.show()


c1.update("Akash",20)
c2.show()

c1.compare(c2)
```
